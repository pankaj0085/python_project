# Exploratary Data Analysis

```python
df.head()
```

```python
graph1 = sns.countplot(x='Gender',data = df)
for bars in graph1.containers:
    graph1.bar_label(bars)
# This graph will show that female customers have purchased more than male customers    
```

```python
sales_gen = df.groupby(['Gender'])['Amount'].sum().reset_index().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales_gen,)
# This graph shows that female customers have spend more than male customers
```

```python
ax = sns.countplot(data=df,x='Age Group',hue='Gender',palette='mako')
for bars in ax.containers:
    ax.bar_label(bars)
# This graph shows that most of the customers are from age group 26-35 and most of them are female    
```

```python
sales_age = df.groupby(['Age Group'])['Amount'].sum().reset_index().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age,palette='mako')

# This graph shows that the age group of 26-35 has purchased the most and most of the customers are females 
```

```python
# Statewise Analysis
sales_state = df.groupby(['State'])['Orders'].sum().reset_index().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,8)})
sns.barplot(x='State',y='Orders',data=sales_state,palette='rocket')
# This graph shows the top 10 states where the most orders were placed
```

```python
# Top 5 states based on sales amount
sales_state_amount= df.groupby(['State'])['Amount'].sum().reset_index().sort_values(by='Amount',ascending=False).head(5)
sns.barplot(x='State',y='Amount',data=sales_state_amount,palette='coolwarm')
```

```python
ax = sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(5,6)})
for bars in ax.containers:
    ax.bar_label(bars)
# This graph shows that most of the customers are married    
```

```python
sales_marital = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().reset_index().sort_values(by='Amount',ascending=False)   
sns.set(rc={'figure.figsize':(7,6)})
sns.barplot(data=sales_marital,x='Marital_Status',y='Amount',hue='Gender',palette='mako')
# This graph shows that married women have purchased the most
```

```python
sns.set(rc={'figure.figsize' : (20,5)})
ax = sns.countplot(data=df,x='Occupation',palette='rocket')
for bars in ax.containers:
    ax.bar_label(bars)
# This graph shows that most costumers are from IT sector    
```

```python
sales_occupation=df.groupby(['Occupation'], as_index=False)['Amount'].sum().reset_index().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,6)})
sns.barplot(data=sales_occupation,x='Occupation',y='Amount',palette='mako')
# This graph shows that most costumers are from IT sector and they have purchased the most
```

```python
sns.set(rc={'figure.figsize' : (26,5)})
ax = sns.countplot(data=df,x='Product_Category',palette='rocket')
for bars in ax.containers:
    ax.bar_label(bars)
# This graph shows that clothing and accessories category has the highest number of orders    
```

```python
sales_category = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().reset_index().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_category,x='Product_Category',y='Amount',palette='Set2')
# Ths graph shows tha food and beverages category has the highest sales amount
```
